```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig1 = plt.figure()
ax1 = fig1.add_subplot(111, aspect='equal')
ax1.add_patch(
    patches.Rectangle(
        (0.1, 0.1),   # (x,y)
        0.5,          # width
        0.5,          # height
    )
)
```

```python
from matplotlib.path import Path
verts = [
    (0., 0.), # left, bottom
    (0., 1.), # left, top
    (1., 1.), # right, top
    (1., 0.), # right, bottom
    (0., 0.), # ignored
    ]

codes = [Path.MOVETO,
         Path.LINETO,
         Path.LINETO,
         Path.LINETO,
         Path.CLOSEPOLY,
         ]

path = Path(verts, codes)

fig = plt.figure()
ax = fig.add_subplot(111)
patch = patches.PathPatch(path, facecolor='orange', lw=2)
ax.add_patch(patch)
ax.set_xlim(-2,2)
ax.set_ylim(-2,2)
```

```python
import numpy as np
new_verts = []
for v in verts:
    #print(v)
    vec = np.array(v)
    vec = vec.transpose()
    shearing_matrix = np.array([[1,.5],[0,1]])
    #print(vec)
    #print(shearing_matrix)
    new_vec = shearing_matrix.dot(vec)
    #print(new_vec)
    new_verts.append(tuple(new_vec))
```

```python
import numpy as np
import math as m
new_verts = []
for v in verts:
    #print(v)
    vec = np.array(v)
    vec = vec.transpose()
    degree = 45
    rad = m.pi*degree/180 
    shearing_matrix = np.array([[m.cos(rad),m.sin(rad)],[-m.sin(rad),m.cos(rad)]])
    #print(vec)
    #print(shearing_matrix)
    new_vec = shearing_matrix.dot(vec)
    #print(new_vec)
    new_verts.append(tuple(new_vec))

path = Path(new_verts, codes)

fig = plt.figure()
ax = fig.add_subplot(111)
patch = patches.PathPatch(path, facecolor='orange', lw=2)
ax.add_patch(patch)
ax.set_xlim(-2,2)
ax.set_ylim(-2,2)
```

```python
import numpy as np
from scipy import misc
```

```python
image_iwan = misc.imread('iwan.png')
```

```python
print(image_iwan.shape)
```

```python
print(image_iwan.dtype)
```

```python
image_iwan_normal = 1.0 * image_iwan[:,:,1] /255
image_iwan_normal
```

```python
image_metta = misc.imread('metta.png')
```

```python
print(image_metta.shape)
```

```python
image_metta_normal = 1.0 * image_metta[:,:,1] /255
image_metta_normal
```

```python
import matplotlib.pyplot as plt
def gambar(img):
    temp_image = np.zeros([img.shape[0],img.shape[1],3])
    temp_image[:,:,0] = img
    temp_image[:,:,1] = img
    temp_image[:,:,2] = img
    plt.imshow(temp_image)
    plt.show()

    
#iwan_normal = np.zeros([300,300,3])
#iwan_normal[:,:,0] = image_iwan_normal
#iwan_normal[:,:,1] = image_iwan_normal
#iwan_normal[:,:,2] = image_iwan_normal

#plt.imshow(iwan_normal)
#plt.show()
gambar(image_iwan_normal)
```

```python
gambar(image_metta_normal)
```

```python
for i in np.arange(0,1.01,0.1):
    print(i,1-i)
```

```python
for i in np.arange(0,1.01,0.01):
    image_out = i * image_iwan_normal + (1-i) * image_metta_normal
    gambar(image_out)
    
```
